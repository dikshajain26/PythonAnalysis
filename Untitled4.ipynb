```python
import pandas as pd
import plotly.express as px

# Read the dataset
data = pd.read_excel('/content/Adidas.xlsx')
data.head()

# Assuming 'Date' is a datetime column
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'])

# Convert 'Date' to string
data['InvoiceDate'] = data['InvoiceDate'].astype(str)

# Create a figure with Plotly Express
fig = px.bar(data, x='Product', y='TotalSales', color='Product', animation_frame='InvoiceDate',
             title='Total Revenue Over Time', labels={'Revenue': 'Total Revenue'})

# Show the figure
fig.show()
```
